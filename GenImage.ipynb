```python
import diffusers
import torch
import json
import os
import PIL
import random

model_id = "sd-legacy/stable-diffusion-v1-5"
pipe = diffusers.StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

new_img_path = "/content/drive/MyDrive/CS682/Data Augmentation Project/ImageNet/diff_img/"
dataset_path = "/content/drive/MyDrive/CS682/Data Augmentation Project/ImageNet/8/"
label_path = os.path.join(dataset_path, "Labels.json")

labels = json.load(open(label_path))
```

```python
for dir in os.listdir(dataset_path):
    if dir in ['train.X3','train.X4']:
        dir_path = os.path.join(dataset_path, f'{dir}/')
        for sub_dir in os.listdir(dir_path):
            prompt = labels[sub_dir] + 'animal'
            i: int = 0
            dat_dir = os.path.join(dir_path, sub_dir)
            for img in os.listdir(dat_dir)[:100]:
                random_float = random.uniform(0.25, 0.75)
                img_path = os.path.join(new_img_path, sub_dir)
                if not os.path.exists(os.path.join(img_path, f'{i}.png')):
                    if not os.path.exists(img_path):
                        print("MAKING DIRECTORY -> ", img_path)
                        os.makedirs(img_path)
                    img_path = os.path.join(img_path, f'{i}.png')
                    init_image = PIL.Image.open(os.path.join(dat_dir, img))
                    print("MAKING IMAGE -> ", img_path)
                    image = pipe(prompt, image = init_image, strength = random_float).images[0]
                    image.save(img_path)
                i += 1
```
